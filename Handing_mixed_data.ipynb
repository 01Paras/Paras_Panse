**Handling Mixed variables**

Here, We have **'Cabin'**,**'Ticket'** and **'number'** variables.







```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/titanic (1).csv')
```

```python
df.head()
```

```python
df['number'].unique()
```

```python
df['number'].value_counts()
```

```python
# to extract numerical part from the value.
df['num_numeric'] = pd.to_numeric(df['number'],errors='coerce',downcast='integer')
```

```python
# Extract the categorical data from the mixed values.
df['num_categorical'] = np.where(df['num_numeric'].isnull(),df['number'],np.nan)
```

```python
df.head()
```

```python
df['Cabin'].unique()
```

```python
df['Ticket'].unique()
```

```python
df['Cabin_num'] = df['Cabin'].str.extract('(\d+)')
df['Cabin_cat'] = df['Cabin'].str[0]
df.head()
```

```python
df['Cabin_cat'].value_counts()
```

```python
df['Ticket'][0].split()
```

```python
df['Ticket_num'] =  df['Ticket'].apply(lambda s:s.split()[-1])
df['Ticket_num'] = pd.to_numeric(df['Ticket_num'],errors='coerce',downcast='integer')

df['Ticket_cat'] =  df['Ticket'].apply(lambda s:s.split()[0])
df['Ticket_cat'] = np.where(df['Ticket_cat'].str.isdigit(),np.nan,df['Ticket_cat'])
df.head(20)
```

```python
print(df['Ticket_cat'].unique())

print(df['Ticket_cat'].value_counts())
```
